```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_N = pd.read_csv('times_N.csv', header = 0, sep=' *, *', engine ='python')
time_N
```

```python
g = sns.lineplot(x="n", y="timeCPU", marker="o", dashes=True, data=time_N, label="CPU time")
g = sns.lineplot(x="n", y="timeGPU", marker="o", dashes=True, data=time_N, label="GPU time")
g.set_title("Increasing elements in matrix proportionally incrising the grid size, constant block size (32x32)")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```
