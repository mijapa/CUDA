```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_N = pd.read_csv('times_threads.csv', header = 0, sep=' *, *')
time_N
```

```python
g = sns.lineplot(x="threadsPerBlock", y="timeCPU", marker="o", dashes=True, data=time_N, label="CPU time")
g = sns.lineplot(x="threadsPerBlock", y="timeGPU", marker="o", dashes=True, data=time_N, label="GPU time")
g.set_title("Increasing the number of threads while proportionally reducing the block size")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="threadsPerBlock", y="blocksPerGrid", marker="o", dashes=True, data=time_N)
g.set_title("Increasing the number of threads while proportionally reducing the block size")

g.figure.set_size_inches(20,8)
```
