```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_N = pd.read_csv('times_threads.csv', header = 0, sep=' *, *', engine='python')
time_N.n=time_N.n.pow(1./2)
time_N
```

```python
g = sns.lineplot(x="threadsPerBlock", y="timeGPU", marker="o", dashes=True, data=time_N, label="GPU time")
g.set_title("Increasing the number of threads while proportionally reducing the block size, constant problem size (512x512)")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```

zwiększanie ilości wątków w bloku przy jednoczesnym zmniejszeniu ilośći bloków pozwala na znaczne przyspieszenie obliczeń

```python
g = sns.lineplot(x="threadsPerBlock", y="blocksPerGrid", marker="o", dashes=True, data=time_N)
g.set_title("Increasing the number of threads while proportionally reducing the block size, constant problem size (512x512)")
g.figure.set_size_inches(20,8)
```
