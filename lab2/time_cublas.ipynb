```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_N = pd.read_csv('times_CUBLAS.csv', header = 0, sep=' *, *', engine='python')
time_N.n=time_N.n.pow(1./2)
time_N
```

```python
g = sns.lineplot(x="n", y="timeCPU", marker="o", dashes=True, data=time_N, label="CPU time")
g = sns.lineplot(x="n", y="timeGPU", marker="o", dashes=True, data=time_N, label="my GPU time")
g = sns.lineplot(x="n", y="timeCUBLAS", marker="o", dashes=True, data=time_N, label="cublas GPU time")
g.set_title("Increasing matrix size, constatn 32 thhreads per block, reducing grid size accordingly")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="n", y="timeGPU", marker="o", dashes=True, data=time_N, label="my GPU time")
g = sns.lineplot(x="n", y="timeCUBLAS", marker="o", dashes=True, data=time_N, label="cublas GPU time")
g.set_title("Increasing matrix size, constatn 32 thhreads per block, reducing grid size accordingly (without CPU)")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="n", y="gflopsCPU", marker="o", dashes=True, data=time_N, label="CPU")
g = sns.lineplot(x="n", y="gflopsGPU", marker="o", dashes=True, data=time_N, label="GPU")
g = sns.lineplot(x="n", y="gflops", marker="o", dashes=True, data=time_N, label="cublas GPU")

g.set_title("Increasing matrix size, constatn 32 thhreads per block, reducing grid size accordingly")
g.set_ylabel("GigaFlops")
g.figure.set_size_inches(20,8)
```
