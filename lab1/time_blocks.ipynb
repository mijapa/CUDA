```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_N = pd.read_csv('times_blocks.csv', header = 0, sep=' *, *')
# time_N
```

```python
g = sns.lineplot(x="blocksPerGrid", y="timeCPU", marker="o", dashes=True, data=time_N, label="CPU time")
g = sns.lineplot(x="blocksPerGrid", y="timeGPU", marker="o", dashes=True, data=time_N, label="GPU time")
g.set_title("Increasing the number of blocks at a constant (1024) number of threads per block")
g.set_ylabel("Time [ms]")
g.figure.set_size_inches(20,8)
```
