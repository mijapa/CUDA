```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_occupancy = pd.read_csv('times_occupancy.csv', header = 0, sep=' *, *', engine ='python')
time_occupancy
```

```python
g = sns.lineplot(x="threadsPerBlock", y="occupancyTheoretical", marker="o", dashes=True, data=time_occupancy, label="theoretical")
g = sns.lineplot(x="threadsPerBlock", y="occupancyAchieved", marker="o", dashes=True, data=time_occupancy, label="achieved")

g.set_title("Increasing the number of threads while proportionally reducing the block size, constant problem size (128)")
g.set_ylabel("Occupancy [%]")
g.figure.set_size_inches(20,8)
```
