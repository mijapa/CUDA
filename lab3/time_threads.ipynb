```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time_threads = pd.read_csv('times_threads.csv', header = 0, sep=' *, *', engine ='python')
time_threads
```

```python
g = sns.lineplot(x="threadsPerBlock", y="timeGPU", marker="o", dashes=True, data=time_threads, hue="type")
g.set_title("Increasing the number of threads while proportionally reducing the block size, constant problem size (128)")
g.set_ylabel("Time [s]")
g.figure.set_size_inches(20,8)
```

```python
g = sns.lineplot(x="threadsPerBlock", y="blocksPerGrid", marker="o", dashes=True, data=time_threads)
g.set_title("Increasing the number of threads while proportionally reducing the block size, constant problem size (128)")
g.figure.set_size_inches(20,8)
```
