```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline


time = pd.read_csv('times.csv', header = 0, sep=' *, *', engine ='python')
time
```

```python
s = time.groupby("type", as_index=False).mean()
g = sns.barplot(s.type, s.timeGPU, palette='rainbow')
g.set_title("Times for different types of memmory")
g.set_ylabel("Time [s]")
g.figure.set_size_inches(20,8)
```
